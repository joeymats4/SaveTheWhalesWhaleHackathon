```python
#Spatial Analysis of Chlorophyll Concentrations across lat-lons
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np
import ipywidgets as widgets
from IPython.display import display

df_cleaned = df_cleaned.dropna(subset=[
    'ChlorA', 'Lat_Dec', 'Lon_Dec', 'Depthm',
    'Salnty', 'T_degC', 'NO3uM', 'O2ml_L', 'SiO3uM'
])
df_cleaned = df_cleaned[df_cleaned['NO3uM'] >= 0]

#Date Selection (MM/YYYY)
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])
df_cleaned['Year'] = df_cleaned['Date'].dt.year
df_cleaned['Month'] = df_cleaned['Date'].dt.month

month_options = [('All', 0)] + [(pd.Timestamp(2000, m, 1).strftime('%B'), m) for m in range(1, 13)]
month_dropdown = widgets.Dropdown(
    options=month_options,
    value=0,
    description='Month:',
    style={'description_width': 'initial'}
)

quarter_options = [
    ('All', 0),
    ('Q1 — Winter (Jan-Mar)', [1, 2, 3]),
    ('Q2 — Spring (Apr-Jun)', [4, 5, 6]),
    ('Q3 — Summer (Jul-Sep)', [7, 8, 9]),
    ('Q4 — Fall (Oct-Dec)', [10, 11, 12])
]
quarter_dropdown = widgets.Dropdown(
    options=quarter_options,
    value=0,
    description='Quarter:',
    style={'description_width': 'initial'}
)

# Year dropdown — 0 = all years
year_options = [('All', 0)] + [(str(y), y) for y in sorted(df_cleaned['Year'].unique())]
year_dropdown = widgets.Dropdown(
    options=year_options,
    value=0,
    description='Year:',
    style={'description_width': 'initial'}
)

#Range Sliders
def make_range_slider(label, column, step=1.0):
    col_min = float(np.floor(df_cleaned[column].min()))
    col_max = float(np.ceil(df_cleaned[column].max()))
    return widgets.FloatRangeSlider(
        value=[col_min, col_max],
        min=col_min,
        max=col_max,
        step=step,
        description=label,
        continuous_update=False,
        style={'description_width': 'initial'},  # prevents label truncation
        layout=widgets.Layout(width='500px')
    )

depth_slider   = make_range_slider('Depth (m)',        'Depthm',  step=10.0)
sal_slider     = make_range_slider('Salinity (PSS)',   'Salnty',  step=0.1)
temp_slider    = make_range_slider('Temp (°C)',        'T_degC',  step=0.5)
nitrate_slider = make_range_slider('Nitrate (µmol/L)', 'NO3uM',  step=0.5)
oxygen_slider  = make_range_slider('Oxygen (mL/L)',    'O2ml_L',  step=0.1)
silicate_slider= make_range_slider('Silicate (µmol/L)','SiO3uM', step=0.5)

def plot_filtered(depth, salinity, temperature, nitrate, oxygen, silicate, month, quarter, year):
    mask = (
        df_cleaned['Depthm'].between(*depth) &
        df_cleaned['Salnty'].between(*salinity) &
        df_cleaned['T_degC'].between(*temperature) &
        df_cleaned['NO3uM'].between(*nitrate) &
        df_cleaned['O2ml_L'].between(*oxygen) &
        df_cleaned['SiO3uM'].between(*silicate)
    )

    if month != 0:
        mask &= df_cleaned['Month'] == month
    if quarter != 0:
        mask &= df_cleaned['Month'].isin(quarter)
    if year != 0:
        mask &= df_cleaned['Year'] == year

    subset = df_cleaned[mask]

    fig, ax = plt.subplots(figsize=(10, 8), subplot_kw={'projection': ccrs.PlateCarree()})

    if subset.empty:
        ax.set_title('No data matches current filter range')
        plt.show()
        return

    sc = ax.scatter(
        subset['Lon_Dec'], subset['Lat_Dec'],
        c=np.log1p(subset['ChlorA']),
        cmap='viridis', s=5,
        vmin=0, vmax=np.log1p(df_cleaned['ChlorA'].max()),
        transform=ccrs.PlateCarree()
    )
    ax.add_feature(cfeature.LAND, zorder=1)
    ax.add_feature(cfeature.COASTLINE, zorder=2)
    ax.set_extent([-130, -110, 24, 42])
    ax.set_title(f'ChlorA — {len(subset)} points after filtering')
    plt.colorbar(sc, ax=ax, label='log(ChlorA + 1) [µg/L]')
    plt.show()

widgets.interact(
    plot_filtered,
    depth=depth_slider,
    salinity=sal_slider,
    temperature=temp_slider,
    nitrate=nitrate_slider,
    oxygen=oxygen_slider,
    silicate=silicate_slider,
    month=month_dropdown,
    quarter=quarter_dropdown,
    year=year_dropdown
)
```

```python
from scipy.stats import binned_statistic

df_cleaned_1 = df_total
df_cleaned_1 = df_cleaned_1.dropna(subset=[
    'ChlorA', 'Lat_Dec', 'Lon_Dec', 'Depthm',
    'Salnty', 'T_degC', 'NO3uM', 'O2ml_L', 'SiO3uM', 'Distance'
])
df_cleaned_1 = df_cleaned_1[df_cleaned_1['NO3uM'] >= 0]

depth_bins_options = [10, 50, 100, 200]

depth_bin_dropdown = widgets.Dropdown(
    options=[(f'0–{d}m', d) for d in depth_bins_options],
    value=10,
    description='Max Depth (m):',
    style={'description_width': 'initial'}
)

month_dropdown_cg = widgets.Dropdown(
    options=[('All', 0)] + [(pd.Timestamp(2000, m, 1).strftime('%B'), m) for m in range(1, 13)],
    value=0,
    description='Month:',
    style={'description_width': 'initial'}
)

def plot_coastal_gradient(max_depth, month):
    subset = df_cleaned_1[df_cleaned_1['Depthm'] <= max_depth].copy()
    if month != 0:
        subset = subset[subset['Month'] == month]

    # Bin by distance
    bin_means, bin_edges, _ = binned_statistic(
        subset['Distance'], subset['ChlorA'], statistic='mean', bins=30
    )
    bin_std, _, _ = binned_statistic(
        subset['Distance'], subset['ChlorA'], statistic='std', bins=30
    )
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2

    fig, ax = plt.subplots(figsize=(10, 5))
    ax.plot(bin_centers, bin_means, color='green', linewidth=2, label='Mean ChlorA')
    ax.fill_between(
        bin_centers,
        bin_means - bin_std,
        bin_means + bin_std,
        alpha=0.3, color='green', label='±1 SD'
    )
    ax.set_xlabel('Distance from Coast (nautical miles)')
    ax.set_ylabel('ChlorA (µg/L)')
    ax.set_title(f'Coastal Distance Gradient — Depth ≤ {max_depth}m')
    ax.legend()
    ax.set_ylim(0, 6)
    plt.tight_layout()
    plt.show()

widgets.interact(plot_coastal_gradient, max_depth=depth_bin_dropdown, month=month_dropdown_cg)
```

```python
nutrient_options = {
    'Nitrate (NO3uM)': 'NO3uM',
    'Phosphate (PO4uM)': 'PO4uM',
    'Silicate (SiO3uM)': 'SiO3uM'
}

nutrient_dropdown = widgets.Dropdown(
    options=list(nutrient_options.keys()),
    value='Nitrate (NO3uM)',
    description='Nutrient:',
    style={'description_width': 'initial'}
)

depth_range_slider = widgets.FloatRangeSlider(
    value=[0, 50],
    min=0,
    max=float(np.ceil(df_cleaned_1['Depthm'].max())),
    step=10,
    description='Depth (m):',
    continuous_update=False,
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='500px')
)

month_dropdown_nut = widgets.Dropdown(
    options=[('All', 0)] + [(pd.Timestamp(2000, m, 1).strftime('%B'), m) for m in range(1, 13)],
    value=0,
    description='Month:',
    style={'description_width': 'initial'}
)

def plot_nutrient_chl(nutrient_label, depth_range, month):
    nutrient_col = nutrient_options[nutrient_label]

    subset = df_cleaned_1[
        df_cleaned_1['Depthm'].between(*depth_range)
    ].dropna(subset=[nutrient_col, 'ChlorA'])

    if month != 0:
        subset = subset[subset['Month'] == month]

    # Bin nutrient into percentiles to show trend cleanly
    subset = subset.copy()
    subset['nutrient_bin'] = pd.qcut(subset[nutrient_col], q=20, duplicates='drop')
    bin_stats = subset.groupby('nutrient_bin')['ChlorA'].agg(['mean', 'std']).reset_index()
    bin_centers = subset.groupby('nutrient_bin')[nutrient_col].mean().values

    fig, axes = plt.subplots(1, 2, figsize=(14, 5))

    # Left: raw scatter
    axes[0].scatter(
        subset[nutrient_col], subset['ChlorA'],
        alpha=0.2, s=5, color='steelblue'
    )
    axes[0].set_xlabel(nutrient_label)
    axes[0].set_ylabel('ChlorA (µg/L)')
    axes[0].set_title('Raw Scatter')

    # Right: binned mean with SD
    axes[1].plot(bin_centers, bin_stats['mean'], color='steelblue', linewidth=2)
    axes[1].fill_between(
        bin_centers,
        bin_stats['mean'] - bin_stats['std'],
        bin_stats['mean'] + bin_stats['std'],
        alpha=0.3, color='steelblue'
    )
    axes[1].set_xlabel(nutrient_label)
    axes[1].set_ylabel('Mean ChlorA (µg/L)')
    axes[1].set_ylim(0, 5)
    axes[1].set_title('Binned Mean ± 1 SD')

    fig.suptitle(f'{nutrient_label} vs ChlorA — Depth {depth_range[0]}–{depth_range[1]}m')
    plt.tight_layout()
    plt.show()

widgets.interact(
    plot_nutrient_chl,
    nutrient_label=nutrient_dropdown,
    depth_range=depth_range_slider,
    month=month_dropdown_nut
)
```
